# Isi Unique Count per Kolom

```python
print("Isi (Nilai Unik) Semua Kolom:")
print("="*60)

for col in df.columns:
    unique_vals = df[col].dropna().unique()
    
    if len(unique_vals) <= 100:
        print(f"Kolom: {col} ({len(unique_vals)} nilai unik)")
        print(f"Isi: {list(unique_vals)}")
        
    else:
        print(f"Kolom: {col} ({len(unique_vals)} nilai unik)")
        print(f"Isi: [Lebih dari 100 variasi, menampilkan 10 sampel...] {list(unique_vals[:16])}")
        print("...")
        
    print("-" * 60)
```

# jumlah per kolom

```python
for col in df.columns:
    if col in ["job_id"]:
        continue
    
    print(f"Kolom: {col} ({df[col].nunique()} nilai unik)")
    print("="*50)
    
    counts = df[col].value_counts()
    pct = df[col].value_counts(normalize=True).mul(100).round(1)
    
    result = pd.DataFrame({
        'Jumlah': counts,
        'Persentase (%)': pct
    })
    
    # Jika nilai unik > 20, tampilkan hanya 10 teratas & 5 terbawah
    if len(result) > 20:
        print(">> Top 10:")
        display(result.head(10))
        print(">> Bottom 5:")
        display(result.tail(5))
    else:
        display(result)
    
    print("\n")
```
