```python
import pandas as pd
import numpy as np
import polars as plt

df = pd.read_csv("ai_impact_jobs_2010_2025.csv")

```

# GROUPING TAHUN

## grouping geografis

negara paling banyak & paling sedikit muncul per tahun

```python
df.groupby('posting_year')['country'].agg(
    jumlah='count',
    paling_banyak=lambda x: x.value_counts().idxmax(),
    paling_sedikit=lambda x: x.value_counts().idxmin(),
    median_negara=lambda x: x.value_counts().sort_values().index[len(x.value_counts()) // 2]
).reset_index()
```

region paling banyak & paling sedikit muncul per tahun

```python
df.groupby('posting_year')['region'].agg(
    jumlah='count',
    paling_banyak=lambda x: x.value_counts().idxmax(),
    paling_sedikit=lambda x: x.value_counts().idxmin(),
    median_negara=lambda x: x.value_counts().sort_values().index[len(x.value_counts()) // 2]
).reset_index()
```
