# load libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px

df = pd.read_csv("ai_impact_jobs_2010_2025.csv")



print("Library imported.")
```

# Visualisation

## jenis-jenis industry

```python
print("Jenis Industry:", df['industry'].unique())
```

## grouping dan visualisasi automation_risk_score berdasarkan industry

```python
stats_auto = df.groupby('industry')['automation_risk_score'].describe()
print("Statistik Automation Risk Score:")
display(stats_auto)

fig1 = px.box(df, x="industry", y="automation_risk_score", 
             title="Distribution of Automation Risk Score by Industry",
             color="industry")
fig1.show()
```

## grouping dan visualisasi AI Job Displacement Risk berdasarkan industry

```python
fig2 = px.pie(df, names='ai_job_displacement_risk', 
             title='Percentage of AI Job Displacement Risk')
fig2.show()

fig3 = px.histogram(
    df, 
    x="industry", 
    color="ai_job_displacement_risk",
    title="AI Job Displacement Risk by Industry",
    category_orders={"ai_job_displacement_risk": ["Low", "Medium", "High"]},
    barmode='stack'
)

fig3.show()
```

## grouping dan visualisasi AI Intensity Score berdasarkan industry

```python
fig4 = px.box(df, x="industry", y="ai_intensity_score", 
             title="AI Intensity Score by Industry",
             color="industry")
fig4.show()
```

## grouping dan visualisasi Industry AI Adoption Stage berdasarkan industry

```python
fig5 = px.pie(df, names='industry_ai_adoption_stage', 
             title='Industry AI Adoption Stage Distribution')
fig5.show()


fig6 = px.histogram(
    df, 
    x="industry", 
    color="industry_ai_adoption_stage",
    title="AI Adoption Stage per Industry",
    # Order from earliest stage to most advanced
    category_orders={"industry_ai_adoption_stage": ["Emerging", "Growing", "Mature"]},
    barmode='stack'
)

fig6.show()
```

## grouping dan visualisasi Salary Box Plot berdasarkan industry

```python
fig7 = px.box(df, x="industry", y="salary_usd", 
             title="Salary Distribution (USD) by Industry",
             color="industry")
fig7.show()
```

## grouping dan visualisasi Reskilling Required berdasarkan industry

```python
# Pie Chart
fig8 = px.pie(df, names='reskilling_required', 
             title='Overall Reskilling Requirement Percentage')
fig8.show()

# Step Bar
fig9 = px.histogram(df, x="industry", color="reskilling_required",
                   title="Reskilling Requirement by Industry",
                   barmode='stack')
fig9.show()
```
